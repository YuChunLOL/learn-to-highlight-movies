### 開始訓練 doc2vec

```python
class LabeledLineSentence(object):
    def __init__(self, sources):
        self.sources = sources
        
        flipped = {}
        
        # make sure that keys are unique
        for key, value in sources.items():
            if value not in flipped:
                flipped[value] = [key]
            else:
                raise Exception('Non-unique prefix encountered')
    
    def __iter__(self):
        for source, prefix in self.sources.items():
            with utils.smart_open(source) as fin:
                for item_no, line in enumerate(fin):
                    yield LabeledSentence(utils.to_unicode(line).split(), [prefix + '_%s' % item_no])
    
    def to_array(self):
        self.sentences = []
        for source, prefix in self.sources.items():
            with utils.smart_open(source) as fin:
                for item_no, line in enumerate(fin):
                    self.sentences.append(LabeledSentence(utils.to_unicode(line).split(), [prefix + '_%s' % item_no]))
        return self.sentences
    
    def sentences_perm(self):
        shuffle(self.sentences)
        return self.sentences
```

```python
sources = {
    'training-data/train_POS.txt':'TRAIN_POS',
    'training-data/train_NEG.txt':'TRAIN_NEG',
    'training-data/test_POS.txt': 'TEST_POS',
    'training-data/test_NEG.txt': 'TEST_NEG'
}

sentences = LabeledLineSentence(sources)
```

```python
model = Doc2Vec(min_count=10, window=20, size=100, sample=1e-5, negative=15, workers=20)
model.build_vocab(sentences.to_array())
```

```python
for epoch in range(20):
    print 'Epoch: {}'.format(epoch+1)
    start_time = time.time()
    model.train(sentences.sentences_perm())
    print '--- {} seconds ---'.format(time.time() - start_time)
```

### 開始 Classification

### 把 training data, testing data 的 vectors, label 都各自放進 array (處理成 scikit-learn model input format)

```python
vars(model) # 給你們看看 doc2vec model 長啥樣子
```

```python
train_arrays = np.zeros((train_size, model.vector_size), dtype='f') # key: 第幾個 document; value: document vector (100-D)
train_labels = np.zeros(train_size) # POS: 1; NEG: 0

for i in range(train_size/2):
    prefix_train_pos = 'TRAIN_POS_' + str(i)
    prefix_train_neg = 'TRAIN_NEG_' + str(i)
    train_arrays[i] = model.docvecs[prefix_train_pos]
    train_arrays[(train_size/2) + i] = model.docvecs[prefix_train_neg]
    train_labels[i] = 1
    train_labels[(train_size/2) + i] = 0
    
test_arrays = np.zeros((test_size, model.vector_size)) # key: 第幾個 document; value: document vector (100-D)
test_labels = np.zeros(test_size) # POS: 1; NEG: 0

for i in range(test_size/2):
    prefix_test_pos = 'TEST_POS_' + str(i)
    prefix_test_neg = 'TEST_NEG_' + str(i)
    test_arrays[i] = model.docvecs[prefix_test_pos]
    test_arrays[(test_size/2) + i] = model.docvecs[prefix_test_neg]
    test_labels[i] = 1
    test_labels[(test_size/2) + i] = 0
```

### 第 0 個 document 的 vector 就是 train_arrays[0]，其對應的 label 就是 train_labels[0]

### Train a classification model (need keras)

```python
from keras.models import Sequential
from keras.models import load_model
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(500, input_dim=100, init='normal', activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(500, init='normal', activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(train_arrays, train_labels,
          nb_epoch=100,
          batch_size=128)
score = model.evaluate(test_arrays, test_labels, batch_size=128)
```
